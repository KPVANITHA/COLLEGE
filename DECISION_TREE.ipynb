```python
import pandas as pd
import math

#returns the column with the highest information gain
def findnode(data):
    max_gain = 0
    array = data.values
    cols = data.columns
    # print(cols)
    info_gain = []

    for j in range(0, len(cols)):
        sub_total_array = []
        sub_entropy_array = []

        col_total = len(array)
        col_pos = 0
        col_neg = 0
        for k in range(0, len(array)):
            #no .of positive rows in the column
            if array[k][-1] == "yes":
                col_pos += 1
            else:#negative rows in the column
                col_neg += 1
        col_total = col_neg + col_pos

        col_entropy = (-(col_pos / col_total) * (math.log2(col_pos / col_total))) - (
            (col_neg / col_total) * (math.log2(col_neg / col_total))
        )

        column = []

        sub_col = []
        children = []
        # checks if the elements of the row are already in the sub_col list.
        #If they are not, then the element is appended to the sub_col list.
        for i in range(0, len(array)):
            if array[i][j] not in sub_col:
                sub_col.append(array[i][j])



        for k in sub_col:
            sub_pos = 0
            sub_neg = 0

            for l in range(0, len(array)):
                if (array[l][j] == k) and ((array[l][-1] == "yes")):
                    sub_pos = sub_pos + 1
                elif (array[l][j] == k) and ((array[l][-1] == "no")):
                    sub_neg = sub_neg + 1

            #it checks if the value in the specified column is equal to the current value of the k variable. If it is,
            # then the function increments the sub_pos v
            if sub_neg == 0:
                children.append([k, "Positive"])
            elif sub_pos == 0:
                children.append([k, "Negative"])
            else:
                children.append([k, "Explore"])
            #################################

            total = sub_pos + sub_neg
            sub_total_array.append(total)
            if sub_neg == 0 or sub_pos == 0:
                sub_entropy = 0
            else:
                sub_entropy = (-(sub_pos / total) * (math.log2(sub_pos / total))) - (
                    (sub_neg / total) * (math.log2(sub_neg / total))
                )
            sub_entropy_array.append(sub_entropy)

        # ig
        sumation = 0
        for m, n in zip(sub_total_array, sub_entropy_array):

            temp = (m / col_total) * n
            sumation = sumation + temp


        gain = col_entropy - (sumation)
        print("IG ( ", cols[j], ") = ", gain)
        info_gain.append(gain)
 #returns the column with the highest information gain.
        for z in info_gain:
            if cols[j] != cols[-1]:
                if max_gain < z:
                    max_gain = z
                    node = cols[j]
                    childs = sub_col
                    child = children

        # print("Information gain: gain(" , cols[j] ,") = " , gain)
    return node, childs, child


tree = []


data = pd.read_csv("tennis.csv")
node, childs, child = findnode(data)
# head
row = []
row.append(node)

# print("\n ROOT NODE :  ", node)
row.append("-->")
for i in child:
    row.append(i)
tree.append(row)

visited = []

for i in range(0, len(tree)):
    if tree[i][0] not in visited:

        visited.append(tree[i][0])
        index = i
        for j in range(1, len(tree[index])):
            row = []
            children = []
            i = 0
            # print("\n***** ", tree[i][j][0], " Dataset*************\n")
            if tree[i][j][1] == "Explore":
                # print(tree[i][j][0])
                row.append(tree[i][j][0])
                visited.append(tree[i][j][0])
                data = pd.read_csv("tennis.csv")
                data = data[data[tree[i][0]] == tree[i][j][0]]
                node, child, childs = findnode(data)
                #print("chosen node : ", node)
                # print("\nChosen node :  ", node)
                row.append(node)
                tree.append(row)
                children.append(node)
                for i in childs:
                    children.append(i)
                tree.append(children)
            else:
                visited.append(tree[i][j][0])

print("\n--------------DECISION TREE----------------")

for i in tree:
    print("\n\nRoot Node :  " , i[0])
    temp = i[0]
    print("\nIntermediate Nodes\n")
    for j in i:
        if j!= temp and j != "-->":
            print(j, sep=" , ")


```
